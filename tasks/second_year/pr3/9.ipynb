```python
seed(random.randint(0,99999))

class Agent:

    def __init__(self, type):
        self.type = type
        self.draw_location()

    def draw_location(self):
        self.location = uniform(0, 100), uniform(0, 100)

    def get_distance(self, other):
        a = (self.location[0] - other.location[0])**2
        b = (self.location[1] - other.location[1])**2
        return sqrt(a + b)

    def is_happy(self, agents):
        distances = []
        for agent in agents:
            if self != agent:
                distance = self.get_distance(agent)
                distances.append((distance, agent))
        # == сортируем == #
        distances.sort()
        # == Определение дистанции == #
        neighbors = [agent for d, agent in distances[:num_neighbors]]
        # == Сколько таких-же как и он сам == #
        num_same_type = sum(self.type == agent.type for agent in neighbors)
        return num_same_type >= require_same_type

    def update(self, agents):
        while not self.is_happy(agents):
            self.draw_location()


def plot_distribution(agents, cycle_num):
    x_values_0, y_values_0 = [], []
    x_values_1, y_values_1 = [], []
    for agent in agents:
        x, y = agent.location
        if agent.type == 0:
            x_values_0.append(x)
            y_values_0.append(y)
        else:
            x_values_1.append(x)
            y_values_1.append(y)
    fig, ax = plt.subplots(figsize=(8, 8))
    plot_args = {'markersize': 8, 'alpha': 0.6}
    ax.set_facecolor('green')
    ax.plot(x_values_0, y_values_0, 'o', markerfacecolor='red', **plot_args)
    ax.plot(x_values_1, y_values_1, 'o', markerfacecolor='black', **plot_args)
    ax.set_title(f'Цикл {cycle_num-1}')
    plt.show()



num_of_type_0 = 30    # Кол-во агентов 0
num_of_type_1 = 10    # Кол-во агентов 1
num_neighbors = 4     # Кол-во агентов, после которого соседями они быть перестают
require_same_type = 5   # Кол-во соседей для завершения пересторения


agents = [Agent(0) for i in range(num_of_type_0)]
agents.extend(Agent(1) for i in range(num_of_type_1))

input1 = int(input())

count = 1
while input1 > count:
    print('Поихали ', count)
    plot_distribution(agents, count)
    count += 1
    no_one_moved = True
    for agent in agents:
        old_location = agent.location
        agent.update(agents)
        if agent.location != old_location:
            no_one_moved = False
    if no_one_moved:
        print('Конец на ', count, ' цикле, все стоят)000)))')
        break

```
